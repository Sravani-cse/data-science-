```python
import pandas as pd
import numpy as np

# Simulate a small dataset
np.random.seed(42)
data = pd.DataFrame({
    'order_id': [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    'category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Clothing', 'Electronics', 'Books', 'Clothing'],
    'region': ['East', 'West', 'East', 'Central', 'West', 'Central', 'East', 'West'],
    'age': np.random.randint(20, 60, 8),
    'sales': np.random.randint(50, 500, 8),
    'cost': np.random.randint(20, 200, 8)
})
data.loc[2, 'region'] = np.nan # Introduce a missing value for cleaning example
data.loc[5, 'sales'] = 650     # Introduce an outlier

# Inspect the dataset
print("\n--- First 5 Rows ---")
print(data.head())
print("\n--- Column Information ---")
print(data.info())
import matplotlib.pyplot as plt
import seaborn as sns

# 3.1 Overview
print("\n--- Basic Sales Statistics ---")
print(data['sales'].describe())

# 3.2 Visualization (Sales Distribution)
plt.figure(figsize=(8, 5))
sns.histplot(data['sales'], kde=True)
plt.title("Distribution of Sales Price")
plt.xlabel("Sales Price (USD)")
plt.show()

# 3.2 Visualization (Sales by Category)
plt.figure(figsize=(8, 5))
sns.barplot(x='category', y='sales', data=data, estimator=np.sum)
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales (USD)")
plt.show()
# 4.1 Handling Missing Values
# Fill missing 'region' with the mode (most frequent value)
data['region'].fillna(data['region'].mode()[0], inplace=True)
print("\n--- Check Missing Values After Cleaning ---")
print(data.isnull().sum())

# 4.2 Feature Engineering: Calculate Profit
data['profit'] = data['sales'] - data['cost']

# 4.3 Feature Engineering: Age Groups
# Create bins for age groups
bins = [20, 30, 40, 50, 60]
labels = ['20-29', '30-39', '40-49', '50-59']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)

print("\n--- Data with new 'profit' and 'age_group' features ---")
print(data.head())
```
