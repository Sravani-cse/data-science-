```python
import pandas as pd

# Load the dataset (Assume 'house_data.csv' is in the current directory)
try:
    data = pd.read_csv("house_data.csv")
except FileNotFoundError:
    print("Simulating a small dataset since 'house_data.csv' wasn't found.")
    data = pd.DataFrame({
        'sqft': [1500, 2200, 1200, 3000, 1800],
        'beds': [3, 4, 2, 5, 3],
        'n_hood': ['East', 'West', 'East', 'Central', 'West'],
        'year': [1990, 2010, 1950, 2020, 2005],
        'price': [250000, 450000, 180000, 750000, 320000]
    })

# Inspect the dataset
print("\n--- First 5 Rows ---")
print(data.head())
print("\n--- Column Information ---")
print(data.info())
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 3.1 Overview (Simple Statistics)
print("\n--- Basic Statistics ---")
print(data.describe())

# 3.2 Visualization (Price Distribution)
plt.figure(figsize=(8, 5))
sns.histplot(data['price'], kde=True)
plt.title("Distribution of Sale Price")
plt.xlabel("Price (USD)")
plt.show()

# 3.2 Visualization (Price by Neighborhood)
plt.figure(figsize=(8, 5))
sns.boxplot(x='n_hood', y='price', data=data)
plt.title("Sale Price by Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Price (USD)")
plt.show()

# 4.1 Feature Engineering (House Age)
current_year = 2024 # Use current year for calculation
data['house_age'] = current_year - data['year']

# Drop the original 'year' column
data.drop('year', axis=1, inplace=True)

print("\n--- Data with new 'house_age' feature ---")
print(data.head())
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Separate features (X) and target variable (y)
X = data.drop('price', axis=1)
y = data['price']

# Identify categorical and numerical features
categorical_features = ['n_hood']
numerical_features = ['sqft', 'beds', 'house_age']

# Create a column transformer for one-hot encoding and passing through numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('num', 'passthrough', numerical_features)
    ])

# Create a pipeline with the preprocessor and a Linear Regression model
model = Pipeline(steps=[('preprocessor', preprocessor),
                      ('regressor', LinearRegression())])

# Fit the model
model.fit(X, y)

print("Model training complete.")
```
